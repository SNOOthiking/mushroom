```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.pipeline import make_pipeline
```

#1 데이터 전처리 (위에서 본 상관관계와 다중공선성을 확인해서 변수제거)

```python
mushrooms = pd.read_csv('C:/ml2/data/mushrooms.csv')
mushrooms.drop('veil-type',axis=1,inplace=True)
mushrooms.drop('veil-color',axis=1,inplace=True)
mushrooms.drop(['cap-shape','cap-color','odor'],axis=1,inplace=True)
```

#2 데이터 분리 

```python
x_data =mushrooms.drop('class',axis=1)
encode = LabelEncoder()
mushrooms['class'] = encode.fit_transform(mushrooms['class'])
y_data =mushrooms[['class']].values
```

```python
x_data
y_data
```

#3 트레인 테스트 셋 스플릿 

```python
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.2,stratify=y_data)
```

#4 컬럼 원핫인코딩 한번에 

```python
from sklearn.compose import make_column_transformer
```

```python
myt = make_column_transformer(  ( OneHotEncoder(), list(x_data.columns.values) ) ,
                              remainder='passthrough')
result = myt.fit_transform( x_train)
result.toarray()[0]
```

```python
result.shape
```

#5 딥러닝

#6 모델에 적용

```python
def createModel():
    dense1 = Dense( input_dim=87,units=32,activation="relu")
    dense2 = Dense( input_dim=32,units=16,activation='relu')
    dense3 = Dense( input_dim=16,units=1,activation='sigmoid')
    
    model = Sequential( [dense1,dense2,dense3])
    model.compile( loss="binary_crossentropy" ,optimizer=Adam(0.1) )
    return model
```

#7 검증 

```python
kcf = KerasClassifier(build_fn= createModel , epochs=1000, verbose=1)
pmodel = make_pipeline( myt, kcf)
pmodel.fit( x_train, y_train)
```

```python
from sklearn.metrics import f1_score
pred = pmodel.predict(x_test)

f1_score(y_test,pred)
```

```python
pred = pmodel.predict(x_train)
f1_score(y_train,pred)
```
